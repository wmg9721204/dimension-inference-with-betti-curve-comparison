```julia
include("unified_plot_scatter_bettis_mean_bettis.jl")

###########################################
d_up = 5
num_seed = 100 #100
###########################
## moderately frequently modified parameters
cutoff_ind_ratio = 0.50
name_ir = "0$(Int(100*cutoff_ind_ratio))"
animal_name = "fly"
###########################################
## path and controlled name for real data
base_path_real = pwd()*"/fly_pos_resp_real_$(name_ir)/fly_data_pos_resp_bettis_dup_$(d_up)"
name_tail_real = "cutoff_1_$(name_ir)"
###########################################
## path and controlled name for ctrl data
base_path_ctrl = pwd()*"/fly_pos_resp_ctrl_$(name_ir)/fly_data_pos_resp_bettis_dup_$(d_up)"
name_tail_ctrl = "cutoff_1_$(name_ir)"
######################################################
label_real = "$(animal_name)"
base_title = "$(animal_name) vs ctrl"
###########################################

## generate all possible controls
Angles = Array{Int,1}(round.(100*collect(range(pi/4, pi, length = 5))))
num_angles = 10
Angles = Array{Int,1}(round.(100*collect(range(pi/4, pi, length = num_angles))))

dir_types = ["angle_$(angle)" for angle in Angles]
pt_types = ["uniform"]
ctrls_all = reshape([(dir_type, pt_type) for pt_type in pt_types, dir_type in dir_types], 
    length(dir_types)*length(pt_types))

# dir_types = ["pos", "unif"]
# pt_types = ["uniform", "sphere"]
# ctrls_all = reshape([(dir_type, pt_type) for pt_type in pt_types, dir_type in dir_types], 
#     length(dir_types)*length(pt_types))
# push!(ctrls_all, ("unif", "de-sitter"))

ctrl_d_to_betti_pts = Dict()
betti_vec_real = zeros(d_up-1)
##################################################################
d_PCA = 2 ## dimension used for PCA projection
for (dir_type, pt_type) in ctrls_all
    ctrl_result = scatter_mean_betti_PCA(d_PCA::Int, 
        base_path_real::String, name_tail_real::String,
        base_path_ctrl::String, name_tail_ctrl::String,
        dir_type::String, pt_type::String,
        num_seed::Int, base_title::String, label_real::String, 3, d_start = 0)
    for d = 2:10
        ctrl_d_to_betti_pts[(dir_type, pt_type, d)] = ctrl_result[1][(dir_type, pt_type, d)]
    end
    betti_vec_real = ctrl_result[2]
end

##################################################################
d_PCA = 3 ## dimension used for PCA projection
for (dir_type, pt_type) in ctrls_all
    scatter_mean_betti_PCA(d_PCA::Int, 
        base_path_real::String, name_tail_real::String,
        base_path_ctrl::String, name_tail_ctrl::String,
        dir_type::String, pt_type::String,
        num_seed::Int, base_title::String, label_real::String)
end
ctrl_d_to_betti_pts
```

```julia
## construct a dictionary
## keys:(dir_type, pt_type), vals: the corresponding point cloud
ctrl_to_betti_pts = Dict()
for (dir_type, pt_type) in ctrls_all
    Pts_ctrl = zeros(d_up-1, 0)
    for d = 2:10
        Pts_ctrl = [Pts_ctrl ctrl_d_to_betti_pts[(dir_type, pt_type, d)]]
    end
    ctrl_to_betti_pts[(dir_type, pt_type)] = Pts_ctrl
end
ctrl_to_betti_pts

using MultivariateStats
Pts_ctrl = zeros(d_up-1, 0)
for (dir_type, pt_type) in ctrls_all
    Pts_ctrl = [Pts_ctrl ctrl_to_betti_pts[(dir_type, pt_type)]]
end
################################################################################
# train a PCA model in 2D
################################################################################
M_2D = fit(PCA, [Pts_ctrl betti_vec_real]; maxoutdim=2, pratio=1.0)
P_2D = projection(M_2D);

# ################################################################################
# # train a PCA model in 3D
# ################################################################################
# M_3D = fit(PCA, [Pts_ctrl betti_vec_real]; maxoutdim=3) #, pratio=0.99)
# P_3D = projection(M_3D);
```

```julia
using ColorSchemes
colors = ColorSchemes.jet
color_inds = [1,3,5,7,9]

########################################################################
using Plots
pyplot()

scatter()
ind = 1
for (dir_type, pt_type) in ctrls_all
    Pts_ctrl_2D = P_2D'*ctrl_to_betti_pts[(dir_type, pt_type)]
    try
        if dir_type[1:5]=="angle"
            angle_ratio = round(parse(Int, dir_type[7:end])/pi)/100
            scatter!(Pts_ctrl_2D[1,:], Pts_ctrl_2D[2,:], 
                markersize = 5, label = "dir: $(angle_ratio) pi, pt: $(pt_type)", 
                markerstrokewidth=0, alpha = 0.8, color = colors[color_inds[ind]])
        else
            scatter!(Pts_ctrl_2D[1,:], Pts_ctrl_2D[2,:], 
                markersize = 5, label = "dir: $(dir_type), pt: $(pt_type)", 
                markerstrokewidth=0, alpha = 0.8, color = colors[color_inds[ind]])
        end
    catch
        scatter!(Pts_ctrl_2D[1,:], Pts_ctrl_2D[2,:], 
                markersize = 5, label = "dir: $(dir_type), pt: $(pt_type)", 
                markerstrokewidth=0, alpha = 0.8, color = colors[color_inds[ind]])
    end
    ind+=1
end
pt_real_2D = P_2D'*betti_vec_real
scatter!([pt_real_2D[1]], [pt_real_2D[2]], 
    markersize = 15, label = "ratc_007", dpi = 300, markerstrokewidth=0, 
    color = :black, alpha = 1)
scatter!(title = "ratc_008 vs controls (PCA, 2D)
    principal ratio: $(round(100*principalratio(M_2D))/100)")
savefig("scatter_ratc_008_vs_ctrls_PCA_2D.png")

#########################################################################
using Plots
plotly()

scatter()
ind = 1
for (dir_type, pt_type) in ctrls_all
    Pts_ctrl_2D = P_2D'*ctrl_to_betti_pts[(dir_type, pt_type)]
    try
        if dir_type[1:5]=="angle"
            angle_ratio = round(parse(Int, dir_type[7:end])/pi)/100
            scatter!(Pts_ctrl_2D[1,:], Pts_ctrl_2D[2,:], 
                markersize = 5, label = "dir: $(angle_ratio) pi, pt: $(pt_type)", 
                markerstrokewidth=0, alpha = 0.8, color = colors[color_inds[ind]])
        else
            scatter!(Pts_ctrl_2D[1,:], Pts_ctrl_2D[2,:], 
                markersize = 5, label = "dir: $(dir_type), pt: $(pt_type)", 
                markerstrokewidth=0, alpha = 0.8, color = colors[color_inds[ind]])
        end
    catch
        scatter!(Pts_ctrl_2D[1,:], Pts_ctrl_2D[2,:], 
                markersize = 5, label = "dir: $(dir_type), pt: $(pt_type)", 
                markerstrokewidth=0, alpha = 0.8, color = colors[color_inds[ind]])
    end
    ind+=1
end
pt_real_2D = P_2D'*betti_vec_real
scatter!([pt_real_2D[1]], [pt_real_2D[2]], 
    markersize = 10, label = "fly data", dpi = 300, markerstrokewidth=0, 
    color = :black, alpha = 1)
scatter!(title = "ratc_008 vs controls (PCA, 2D)
    principal ratio: $(round(100*principalratio(M_2D))/100)")
savefig("scatter_ratc_008_vs_ctrls_PCA_2D.html")

# #########################################################################
# using Plots
# plotly()

# scatter()
# ind = 1
# for (dir_type, pt_type) in ctrls_all
#     Pts_ctrl_3D = P_3D'*ctrl_to_betti_pts[(dir_type, pt_type)]
#     try
#         if dir_type[1:5]=="angle"
#             angle_ratio = round(parse(Int, dir_type[7:end])/pi)/100
#             scatter!(Pts_ctrl_3D[1,:], Pts_ctrl_3D[2,:], Pts_ctrl_3D[3,:], 
#                 markersize = 1, label = "dir: $(angle_ratio) pi, pt: $(pt_type)", 
#                 markerstrokewidth=0, alpha = 0.8, color = colors[color_inds[ind]])
#         else
#             scatter!(Pts_ctrl_3D[1,:], Pts_ctrl_3D[2,:], Pts_ctrl_3D[3,:], 
#                 markersize = 1, label = "dir: $(dir_type), pt: $(pt_type)", 
#                 markerstrokewidth=0, alpha = 0.8, color = colors[color_inds[ind]])
#         end
#     catch
#         scatter!(Pts_ctrl_3D[1,:], Pts_ctrl_3D[2,:], Pts_ctrl_3D[3,:], 
#                 markersize = 1, label = "dir: $(dir_type), pt: $(pt_type)", 
#                 markerstrokewidth=0, alpha = 0.8, color = colors[color_inds[ind]])
#     end    
#     ind+=1
# end
# pt_real_3D = P_3D'*betti_vec_real
# scatter!([pt_real_3D[1]], [pt_real_3D[2]], [pt_real_3D[3]], 
#     markersize = 5, label = "fly data", dpi = 300, markerstrokewidth=0, 
#     color = :black, alpha = 1)
# scatter!(title = "fly vs controls (PCA, 3D)
#     principal ratio: $(round(100*principalratio(M_3D))/100)")
# savefig("scatter_fly_vs_ctrls_PCA_3D.html")
```
