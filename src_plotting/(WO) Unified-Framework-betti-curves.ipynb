```julia
include("unified_plot_scatter_bettis_mean_bettis.jl")

using Plots
pyplot()

###########################
## less frequently modified parameters
d_start = 1
d_up = 5
num_seed = 100 #100
###########################
## moderately frequently modified parameters
cutoff_ind_ratio = 0.50
name_ir = "0$(Int(100*cutoff_ind_ratio))"
animal_name = "fly"
###########################
###########################################
## parameters to plot the betti curves of real data
###########################################
path_base_real = pwd()*"/fly_pos_resp_real_$(name_ir)/fly_data_pos_resp_bettis_dup_$(d_up)"
name_tail_real = "cutoff_1_$(name_ir)"

###########################################
## parameters to plot the betti curves of controlled data
###########################################
## generate all possible controls
num_angles = 10
Angles = Array{Int,1}(round.(100*collect(range(pi/4, pi, length = num_angles))))

dir_types = ["angle_$(angle)" for angle in Angles]
pt_types = ["uniform"]
ctrls_all = reshape([(dir_type, pt_type) for pt_type in pt_types, dir_type in dir_types], 
    num_angles)

base_data_name = "fly"
for (dir_type, pt_type) in ctrls_all
    for d = 2:10    
        plot()
        plot_betti_curves_real(path_base_real, name_tail_real, 
            base_data_name, d_up, d_start)

        for seed = 1:num_seed
            plot_betti_curves_ctrl(base_path_ctrl::String, name_tail_ctrl::String, 
                d::Int, seed::Int, dir_type::String, pt_type::String,
                base_data_name,
                d_up::Int, d_start)
        end
        plot!()

        ## set the plotting layout
        dpi, ylim, legend = 300, (0,170), :topleft
        plot_betti_layout(dpi, ylim, legend)

        savefig("$(base_data_name)_bettis_dir_$(dir_type)_pt_$(pt_type)_d_$(d)_$(name_ir).png")
    end
end
```
